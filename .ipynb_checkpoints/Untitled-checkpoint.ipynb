```python
def plot_images(images, masks):
    '''
    plot images, masks, and combined images and masks.
    
    params:
    images: ndarray
    masks: ndarray
    '''
    for i in range(len(images)):
        image, mask = images[i], masks[i]
        
        # original image visualization
        fig, ax = plt.subplots(1,3,figsize = (13,11))
        ax[0].imshow(image, cmap = 'gray')
        ax[0].axis('off')

        # mask visualization
        ax[1].imshow(mask, cmap = 'gray')
        ax[1].axis('off')

        # draw mask on top of original image
        ax[2].imshow(image, cmap = 'gray', interpolation = 'none')
        ax[2].imshow(mask, cmap = 'jet', interpolation = 'none', alpha = 0.7)
        ax[2].axis('off')


'''
train image and mask visualization.
slice 72 in the train image and mask has better/full view.
'''
train_mask = nibabel.load(f'{ds_path}/ircad_e05_liver.nii.gz')
train_mask = train_mask.get_fdata()

train_image = nibabel.load(f'{ds_path}/ircad_e05_orig.nii.gz')
train_image = train_image.get_fdata()

# pick a slice (72)
image = train_image[:, :, 72]
mask = train_mask[:, :, 72]

# plot an image, mask, and combined
plot_images([image], [mask])
plot_images([image], [mask])
```

```python
from skimage import io

ROOT = '/Users/gopalcbala/Desktop/Jupyter_NB_Projects/PROJECTS/IRCAD/Untitled/ct-images-semantic-segmentation'
ds_path = f'{ROOT}/ircad-dataset'

def apply_dataset_masking():
    """
    split up original image and mask and store them separately
    """
    # load all original image gz files
    orig_gz_files = [f for f in os.listdir(ds_path) if '_orig' in f]
    print(f'total loaded image file: {len(orig_gz_files)}')

    # traverse and store images and masks
    print('INFO: splitting images and masks..')
    for file in orig_gz_files:
        print(f'INFO: processing - {file}')
        images = nibabel.load(f'{ds_path}/ircad_e05_orig.nii.gz')
        images = images.get_fdata()

        common_name = file.split('.')[0].replace('orig', '')
        masks = nibabel.load(f'{ds_path}/{common_name}liver.nii.gz')
        masks = masks.get_fdata()

        for i in range(images.shape[2]):
            image = images[:, :, i]
            mask = masks[:, :, i]

            image_path = f'{ROOT}/dataset/images_and_masks/{file.split('.')[0]}.jpg'
            mask_path = f'{ROOT}/dataset/images_and_masks/{common_name}liver_mask.jpg'
            pimage.fromarray(np.uint8(image)).save(image_path)
            pimage.fromarray(np.uint8(mask)).save(mask_path)

    print('INFO: processing complete')
    return True

apply_dataset_masking()
```
